```python
import pandas as pd
import numpy as np
```

```python
fd = pd.read_csv('flood_risk_dataset_india.csv')
fd.head(10)
```

```python
floodedAreas = fd[fd['Flood Occurred']>0]
floodedAreas
```

```python
floodedAreas.to_csv('floodedAreas.csv', index=False)
```

```python
NoWaterBody = floodedAreas[floodedAreas['Land Cover']!='Water Body']
NoWaterBody
```

```python
NoWaterBody.to_csv('NoWaterBody.csv', index=False)
```

```python
NoWaterBody.head(20)
```

```python
from matplotlib import pyplot as plt
```

```python
x = NoWaterBody['Land Cover']
y = NoWaterBody['Water Level (m)'] 
plt.xlabel("Land Cover")
plt.ylabel("Water Level")

plt.bar(x, y)
```
